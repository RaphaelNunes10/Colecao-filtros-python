```python
from PIL import Image
from matplotlib import pyplot as plt
import requests

img = Image.open(requests.get("https://www.biography.com/.image/t_share/MTY2MzU3Nzk2OTM2MjMwNTkx/elon_musk_royal_society.jpg", stream=True).raw)
img = img.convert("L")
plt.imshow(img)
```

```python
histograma = img.histogram()
plt.bar(x=range(len(histograma)),height=histograma)
```

```python
# Atributos da Imagem:
# Pendendo para High-key
# Ligeiramente super exposta
# Contraste m√©dio
```

```python
ref=[]
lut=[]
for i in range(256):
    ref.append(i)
    if i > 127:
        val = i*1.4
    else:
        val = i/1.4
    if val>255: val=255 # limite m√°ximo
    if val<0: val=0 # limite m√≠nimo
    lut.append(val)
plt.plot(ref,color='k',linestyle='dotted') # referencia
plt.plot(lut) # look-up table
```

```python
img2 = img.point(lut)
plt.imshow(img2)
```

```python
histograma = img2.histogram()
plt.bar(x=range(len(histograma)),height=histograma)
```
